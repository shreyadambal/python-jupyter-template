### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define x values from 0 to 10
x = np.linspace(0, 10, 100)
# Compute y values as the square of x
y = x**2

# Create the plot
plt.plot(x, y, label='y = x^2', color='b')

# Add title and labels
plt.title('Line Plot of y = x^2')
plt.xlabel('X values')
plt.ylabel('Y values')

# Add grid
plt.grid(True, linestyle='--', alpha=0.7)

# Show legend
plt.legend()

# Display the plot
plt.show()
```
